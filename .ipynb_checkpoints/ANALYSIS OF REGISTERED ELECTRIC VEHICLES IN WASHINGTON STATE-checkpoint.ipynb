```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pySankey.sankey import sankey
from highlight_text import ax_text,fig_text
from pyfonts import load_font
import matplotlib.gridspec as gridspec
```

```python
data=pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

```python
data.shape
```

```python
data["State"].value_counts()
```

```python
data=data.query("State == 'WA'") # We filtered the rows that have 'WA' because we will only focus on cars 
                                 # that are in Washington state.
```

```python
data.isna().any() # There are null values only in the Vehicle Location. 
                  # There is no problem since the variable will not be used.
```

```python
# The Electric Range variable, which was in miles format, has been converted to kilometers format.

data["Electric Range"] = data["Electric Range"].map(lambda x:np.round(x*1.60934,2)) 
```

```python
font = load_font('https://github.com/coreyhu/Urbanist/blob/main/fonts/ttf/Urbanist-Light.ttf?raw=true')
bold_font = load_font('https://github.com/coreyhu/Urbanist/blob/main/fonts/ttf/Urbanist-Medium.ttf?raw=true')
```

# ******

```python
datau=data[["Make","Model"]].value_counts().reset_index()[:15]
datau.columns=["Make","Model","Count_of_Models"]
```

```python
models=data[data["Electric Vehicle Type"] == "Battery Electric Vehicle (BEV)"]["Model"].unique()
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=150)

ax.axis('off') 
table = ax.table(cellText=datau.values, colLabels=datau.columns, cellLoc='center', loc='center')


table.scale(1, 1.5)
table.auto_set_font_size(False)
table.set_fontsize(10)

for i in range(len(datau)):
    for j in range(len(datau.columns)):
        if datau["Model"].iloc[i] in models:
            table[(i + 1, j)].set_facecolor((1.0, 0.0, 0.0, 0.5)) 
        else:
            table[(i + 1, j)].set_facecolor((1.0, 0.678, 0.29, 0.5)) 
for j in range(len(datau.columns)):
    table[(0 , j)].set_facecolor((0.729, 0.643, 0.537, 0.5))  
    
plt.tight_layout()
plt.show()
```

# ******

```python
datap=data["Electric Vehicle Type"].value_counts().reset_index()
change={"Battery Electric Vehicle (BEV)":"BEV","Plug-in Hybrid Electric Vehicle (PHEV)":"PHEV"}
datap["Electric Vehicle Type"]=datap["Electric Vehicle Type"].map(change)
datap
```

```python
fig,ax=plt.subplots(figsize=(5,3),dpi=150)
fig.patch.set_facecolor('#F1F9FA')
ax.set_facecolor('#F1F9FA')

sns.barplot(data=datap,x="Electric Vehicle Type",y="count",palette=["#F13E3E","#FFAD4A"])
sns.despine(ax=ax,top=True,right=True,left=True,bottom=True)

for i, (electric_type, count) in enumerate(zip(datap["Electric Vehicle Type"], datap["count"])):
    ax.text(i, 5000, str(count), ha='center', va='bottom', rotation=90,c="white",fontweight="bold") 
    
ax.set_title(f"Total Number of Registered \nElectric Vehicle Types")
```

# ******

```python
counties=data.County.value_counts()[:11].reset_index()["County"] # The 10 most frequently occurring counties.
counties
```

```python
datao=data[["County","Electric Vehicle Type"]].value_counts().reset_index()
```

```python
datao=datao[datao["County"].isin(counties)]
datao
```

```python
fig,ax = plt.subplots(figsize=(6,4),dpi=150)
fig.patch.set_facecolor('#F1F9FA')
ax.set_facecolor('#F1F9FA')

sns.barplot(x="County",y="count",hue="Electric Vehicle Type", data=datao,palette=["#F13E3E","#FFAD4A"],ci=None,ax=ax)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.legend(frameon=False, loc='upper right', bbox_to_anchor=(1, 0.9))
sns.despine(ax=ax,top=True,right=True,left=True,bottom=True)

ax.set_title("Number of Registered Electric Vehicle Types in Counties (Top 10 Counties)")
```

# ******

```python
makes=data["Make"].value_counts()[:10].reset_index()["Make"]
datax=data[data["Make"].isin(makes)]
```

```python
sankey(datax["Make"],datax["Electric Vehicle Type"],aspect=20, fontsize=12)

fig = plt.gcf()
fig.set_size_inches(15,10)

fig.patch.set_facecolor('#F1F9FA')
plt.title("Types of Vehicles by Registered Make")
```

# ******

```python
datat=data["Make"].value_counts()[:10].reset_index().sort_values(by="count")
datat
```

```python
fig,ax=plt.subplots()
fig.patch.set_facecolor('#F1F9FA')
ax.set_facecolor('#F1F9FA')

colors=sns.color_palette("Spectral",len(datat))

plt.barh(data=datat,y="Make",width="count",color=colors)

for i, (make, count) in enumerate(zip(datat["Make"], datat["count"])):
    ax.text(count + 1000, i, str(count), va='center')
    
sns.despine(ax=ax,top=True, right=True, left=True, bottom=True)
ax.set_title("Makes of Registered Electric Vehicles")
```

# ******

```python
dataz = data[data["Electric Range"] != 0]
```

```python
labelss={"Battery Electric Vehicle (BEV)":"BEV","Plug-in Hybrid Electric Vehicle (PHEV)":"PHEV"}
dataz["Electric Vehicle Type"]=dataz["Electric Vehicle Type"].map(labelss)
```

```python
fig,ax = plt.subplots(figsize=(6,4))

fig.patch.set_facecolor('#F1F9FA')
ax.set_facecolor('#F1F9FA')

sns.set_style('white')

custom_palette = {"BEV": "#F13E3E", "PHEV": "#FFAD4A"}  

sns.violinplot(data=dataz, x="Electric Vehicle Type", y="Electric Range", dodge=False,
                    palette=custom_palette, 
                    scale="width", inner=None,ax=ax)

xlim = ax.get_xlim()
ylim = ax.get_ylim()

for violin in ax.collections:
    bbox = violin.get_paths()[0].get_extents()
    x0, y0, width, height = bbox.bounds
    violin.set_clip_path(plt.Rectangle((x0, y0), width / 2, height, transform=ax.transData))


sns.boxplot(data=dataz, x="Electric Vehicle Type", y="Electric Range",
            saturation=1, showfliers=False, width=0.3,
            boxprops={'zorder': 3, 'facecolor': 'none'}, ax=ax)


old_len_collections = len(ax.collections)

sns.scatterplot(data=dataz, x="Electric Vehicle Type", y="Electric Range",
                hue="Electric Vehicle Type", palette=custom_palette,  
                ax=ax, alpha=0.6)

for dots in ax.collections[old_len_collections:]:
    dots.set_offsets(dots.get_offsets() + np.array([0.12, 0]))

    
ax.grid(alpha=0.6,lw=1.5)      
sns.despine(ax=ax,top=True, right=True, left=True, bottom=True)

ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.legend().remove()

ax.set_title("Electric Range (KM) by Vehicle Type")
```

```python
datab=data[["Make","Electric Vehicle Type"]].value_counts().reset_index()
datab=datab[datab["Make"].isin(makes)]
datab
```

```python
fig,ax = plt.subplots(figsize=(6,4),dpi=150)

fig.patch.set_facecolor('#F1F9FA')
ax.set_facecolor('#F1F9FA')

sns.barplot(x="Make",y="count",hue="Electric Vehicle Type", data=datab,palette=["#F13E3E","#FFAD4A"],ci=None,ax=ax)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

ax.legend(frameon=False, loc='upper right', bbox_to_anchor=(1, 0.9))
ax.set_yticks(np.arange(0,90000,7500))
ax.grid(True, axis='y',alpha=0.3)

sns.despine(ax=ax,top=True,right=True,left=True,bottom=True)

ax.set_title("Types of Vehicles by Registered Make")
```

```python
fig = plt.figure(figsize=(17,12),constrained_layout=True,dpi=150)
gs = fig.add_gridspec(nrows=2,ncols=6)



fig.suptitle(' ANALYSIS OF REGISTERED ELECTRIC VEHICLES IN WASHINGTON STATE ', fontsize=30,font=font)



ax1 = fig.add_subplot(gs[0,0])

fig.patch.set_facecolor('#F1F9FA')
ax1.set_facecolor('#F1F9FA')

sns.barplot(data=datap,x="Electric Vehicle Type",y="count",palette=["#F13E3E","#FFAD4A"],ax=ax1)
sns.despine(ax=ax1,top=True,right=True,left=True,bottom=True)

for i, (electric_type, count) in enumerate(zip(datap["Electric Vehicle Type"], datap["count"])):
    ax1.text(i, count, str(count), ha='center', va='bottom') 

ax1.set_title(f"Total Number of Registered \nElectric Vehicle Types",fontdict={"fontsize":13})
ax1.set_xlabel('')
ax1.set_ylabel('')
fig.tight_layout()


ax2 = fig.add_subplot(gs[0,1:4])
ax2.set_facecolor('#F1F9FA')

sns.barplot(x="County",y="count",hue="Electric Vehicle Type", data=datao,palette=["#F13E3E","#FFAD4A"],ci=None,ax=ax2)

ax2.legend(frameon=False, loc='upper right', bbox_to_anchor=(1, 0.94),fontsize=13)
sns.despine(ax=ax2,top=True,right=True,left=True,bottom=True)
ax2.grid(True, axis='y',alpha=0.5)

ax2.set_title("Number of Registered Electric Vehicle Types in Counties (Top 10 Counties)",fontdict={"fontsize":13})
ax2.set_xlabel('')
ax2.set_ylabel('')
fig.tight_layout()


ax3 = fig.add_subplot(gs[1,0])
ax3.set_facecolor('#F1F9FA')
colors=sns.color_palette("Spectral",len(datat))

ax3.barh(data=datat,y="Make",width="count",color=colors)

for i, (make, count) in enumerate(zip(datat["Make"], datat["count"])):
    ax3.text(count + 1000, i, str(count), va='center')
    
sns.despine(ax=ax3,top=True, right=True, left=True, bottom=True)
ax3.set_title("Makes of Registered Electric Vehicles")
fig.tight_layout()



ax4 = fig.add_subplot(gs[0,4:6])
ax4.set_facecolor('#F1F9FA')

sns.set_style('white')
custom_palette = {'BEV': "#F13E3E", 'PHEV': "#FFAD4A"}  

sns.violinplot(data=dataz, x="Electric Vehicle Type", y="Electric Range", dodge=False,
                    palette=custom_palette,  
                    scale="width", inner=None,ax=ax4)

xlim = ax4.get_xlim()
ylim = ax4.get_ylim()

for violin in ax4.collections:
    bbox = violin.get_paths()[0].get_extents()
    x0, y0, width, height = bbox.bounds
    violin.set_clip_path(plt.Rectangle((x0, y0), width / 2, height, transform=ax4.transData))

sns.boxplot(data=dataz, x="Electric Vehicle Type", y="Electric Range",
            saturation=1, showfliers=False, width=0.3,
            boxprops={'zorder': 3, 'facecolor': 'none'}, ax=ax4)

old_len_collections = len(ax4.collections)

sns.scatterplot(data=dataz, x="Electric Vehicle Type", y="Electric Range",
                hue="Electric Vehicle Type", palette=custom_palette, 
                ax=ax4, alpha=0.6)

for dots in ax4.collections[old_len_collections:]:
    dots.set_offsets(dots.get_offsets() + np.array([0.12, 0]))

    
ax4.grid(alpha=0.6,lw=1.5)      
sns.despine(ax=ax4,top=True, right=True, left=True, bottom=True)

ax4.set_xlim(xlim)
ax4.set_ylim(ylim)
ax4.legend().remove()
ax4.set_xlabel('')
ax4.set_ylabel('')
ax4.set_title("Electric Range (KM) by Vehicle Type")
fig.tight_layout()


ax5 = fig.add_subplot(gs[1,1:4])
ax5.set_facecolor('#F1F9FA')
ax5.set_facecolor('#F1F9FA')

sns.barplot(x="Make",y="count",hue="Electric Vehicle Type", data=datab,palette=["#F13E3E","#FFAD4A"],ci=None,ax=ax5)

ax5.set_xticklabels(ax5.get_xticklabels(), rotation=45)
ax5.set_yticks(np.arange(0,90000,7500))

ax5.grid(True, axis='y',alpha=0.5)
ax5.legend().set_visible(False)
sns.despine(ax=ax5,top=True,right=True,left=True,bottom=True)

ax5.set_title("Types of Vehicles by Registered Make")
ax5.set_xlabel('')
ax5.set_ylabel('')
fig.tight_layout()


ax6 = fig.add_subplot(gs[1, 4:6])
ax6.axis('off')  
table = ax6.table(cellText=datau.values, colLabels=datau.columns, cellLoc='center', loc='center')

table.scale(1, 1.5)
table.auto_set_font_size(False)
table.set_fontsize(10)

for i in range(len(datau)):
    for j in range(len(datau.columns)):
        if datau["Model"].iloc[i] in models:
            table[(i + 1, j)].set_facecolor((1.0, 0.0, 0.0, 0.5)) 
        else:
            table[(i + 1, j)].set_facecolor((1.0, 0.678, 0.29, 0.5)) 
            
for j in range(len(datau.columns)):
    table[(0 , j)].set_facecolor((0.729, 0.643, 0.537, 0.5))   
            

for key, cell in table.get_celld().items():
    cell.set_edgecolor('w')  
    cell.set_text_props(color='black') 

ax6.set_title("Most Registered Models")
plt.tight_layout()

s = '<Graph>: melihakbaba\n<Data Source>: data.wa.gov/Transportation/Electric-Vehicle-Population-Data/f6w7-q2d2/about_data'
fig_text(x=0.05, y=0.035,
         s=s,
        highlight_textprops=[{'font': bold_font}]*2)


fig.savefig("ANALYSIS OF REGISTERED ELECTRIC VEHICLES IN WASHINGTON STATE.png",dpi=200)
```
